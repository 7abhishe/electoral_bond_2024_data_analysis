```python
# prompt: how to read csv file

import pandas as pd

# Read a CSV file
df = pd.read_csv('purcaser.csv')
df1=pd.read_csv('encashment.csv')


# Print the first few rows of the DataFrame
print(df.head())
```

```python
# prompt: now i want unique name of the column name df1 coulumn name of political party and sum  value of denomination

# Get the unique values from the "Political Party" column in df1
unique_parties = df1["Name of the Political Party"].unique()

# Sum the values in the "Denomination" column for each unique party
party_totals = {}
for party in unique_parties:
  party_totals[party] = df1[df1["Name of the Political Party"] == party]["Denomination"].sum()

df1.sort_values(by=['Denomination'], inplace=True, ascending=False)
# Print the results
for party, total in party_totals.items():
  print(f"{party}: {total}")


```
